```python
Q1. What is MongoDB? Explain non-relational databases in short. In which scenarios it is preferred to use
MongoDB over SQL databases?
Ans->
1) MongoDB:
    a) MongoDB is a popular document-oriented NoSQL database that is designed to store, manage and query unstructured or semi-structured data. 
    b) It is an open-source, highly scalable, high-performance, and flexible database that can handle large amounts of data with ease. 
    c) MongoDB stores data in BSON (Binary JSON) format, which is a binary-encoded serialization of JSON-like documents.

2) Non-Relational databases:
    a) Non-relational databases, also known as NoSQL databases, are designed to handle unstructured data, which cannot be easily organized into tables, rows, and columns like relational databases.
    b) Instead of using a fixed schema, NoSQL databases allow users to store and retrieve data in flexible, schema-free data models, such as key-value pairs, documents, graphs, or columns.

3) MongoDB is preferred over SQL databases in scenarios where:
    a) Handling large volumes of unstructured or semi-structured data.
    B) High scalability and performance: MongoDB is highly scalable and can handle a large number of concurrent users and transactions without compromising on performance.
```

```python
# Q5. Explain how you can use the find() method to query the MongoDB database. Write a simple code to
# demonstrate this.
# Ans->
# 1) In MongoDB, the find() method is used to retrieve data from a collection. It accepts a query object as its parameter and returns a cursor object that can be used to iterate over the result set.
# 2) example:
for i in collection.find({'age':{'$gte':18}}):
    print(i)
```

```python
# Q6. Explain the sort() method. Give an example to demonstrate sorting in MongoDB.
# Ans->
# 1) In MongoDB, the sort() method is used to sort the result set of a query in ascending or descending order.
# 2) It accepts a sort object as its parameter, which specifies the fields to sort on and the sorting order.
# 3) example:
s = [('name', -1), ('age', -1)]
cursor = collection.find().sort(s)

for d in cursor:
    print(d)
```

```python
Q7. Explain why delete_one(), delete_many(), and drop() is used.
Ans->
1) delete_one():
    a) This method is used to delete a single document that matches a specified filter.
    b) If multiple documents match the filter, only the first one encountered will be deleted.
2) delete_many():
    a) This method is used to delete all documents that match a specified filter.
    b) If no documents match the filter, nothing will be deleted.
3) drop():
    a) This method is used to drop an entire collection from a database.
    b) This removes all documents from the collection and deletes the collection schema, but does not delete the database itself.
```
