```python
# Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a
# scatter plot for age and fare columns in the titanic dataset.
import seaborn as sb
a = sb.load_dataset('titanic')
fig = go.Figure(data = go.Scatter(x = a.age,y = a.fare))
fig.show()
```

```python
# Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.
import plotly.express as px
tips = px.data.tips()
fig = px.box(tips, x="tip", points="all")
fig.show()
```

```python
from ipywidgets.widgets.widget_int import Color
# Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
# the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
# column with the color parameter.
import plotly.express as px
import seaborn as sb

tips = sb.load_dataset('tips')
fig = px.violin(tips, x='sex', y='total_bill', color='day', box=True, points="all")
fig.show()
```

```python
# Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for
# the color parameter.
# Note: Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the
# dimensions parameter.
import plotly.express as px
from sklearn.datasets import load_iris

iris = load_iris()
df = px.data.iris()

fig = px.scatter_matrix(df, dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"], color="species")
fig.show()
```

```python
# Q5. What is Distplot? Using Plotly express, plot a distplot.
# Ans-> 1) distplot is a histogram with a line plot of the probability density function. 
# 2) It is a useful visualization tool for understanding the distribution of a variable.
import plotly.express as px
tips = px.data.tips()
fig = px.histogram(tips, x="total_bill", nbins=20, opacity=0.5, marginal="box", color_discrete_sequence=['orange'])
fig.show()
```
