```python
x = np.random.rand(50)
y= np.random.rand(50)
plt.scatter(x,y)
```

```python
colours = np.random.rand(50)
plt.scatter(x,y,c=colours)
```

```python
sizes = np.random.rand(50)
plt.scatter(x,y,c=colours,s= sizes) # c  represents the change in color, and s represent size
```

```python
sizes = 1000*np.random.rand(50)
plt.scatter(x,y,c=colours,s= sizes)
```

```python
#alpha represents the intensity
plt.scatter(x,y,c=colours,s= sizes,alpha=.2)
plt.xlabel("the X data")
plt.ylabel("the Y data")
plt.title("THis is my title")
```
