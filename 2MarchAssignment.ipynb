```python
# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
# plot a scatter plot.
# Note: Also add title, xlabel, and ylabel to the plot.
# Ans->
# 1) A scatter plot is a type of plot that displays the relationship between two numerical variables.
# 2) It consists of a set of points, where each point represents a single observation in the data, 
# with the x-coordinate and y-coordinate of each point determined by the values of the two variables being plotted.
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
```

```python
# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
# Use the following data:
# import numpy as np
# For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
# For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
# For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
# For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
# Ans->
# 1) The subplot() function is used to create multiple plots in the same figure.
# 2) It allows you to arrange plots in a grid of rows and columns, and specify which grid cell each plot should occupy.
# 3) This is useful when you want to compare multiple datasets or visualize multiple aspects of the same data in a single figure.


import numpy as np
import matplotlib.pyplot as plt
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
fig,axs = plt.subplots(2, 2)
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')
plt.subplots_adjust(wspace=0.3, hspace=0.3)
plt.show()
```

```python
# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
# Ans-> 1) A bar plot is a way to visualize categorical data with rectangular bars representing the values of each category.
# 2) Bar plots are used to compare different categories based on their values, typically represented on the y-axis, and can show the distribution, frequency, or average of the data.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
```

```python
# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
# Ans-> 1) A box plot is a graphical representation of the distribution of a dataset through its quartiles.
# 2) It shows the minimum and maximum values, the first and third quartiles, and a line for the median of the data.
# 3) It is used to compare the distribution of multiple datasets or to identify the presence of outliers in a dataset.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot([box1,box2])
plt.show()
```
