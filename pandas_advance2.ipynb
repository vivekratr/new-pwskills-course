```python
data1 = pd.DataFrame({'key1':[1,2,4,5,6],
                      'key2':[4,5,6,7,8],
                      'key3':[3,4,5,6,6]
}
)
data2 = pd.DataFrame({'key1':[1,2,45,6,67],
                      'key4':[56,5,6,7,8],
                      'key5':[3,56,5,6,6]
}
)
data1,data2
```

```python
#merge is doing inner join /natural join, using similar column name such as key1 to map both tables
pd.merge(data1,data2) 
```

```python
pd.merge(data1,data2,how='left')  #left outer join
```

```python
pd.merge(data1,data2,how='right')  #right outer join
```

```python
pd.merge(data1,data2,how='outer')  #full outer join
```

```python
pd.merge(data1,data2,how='cross')  #fcross join
```

```python
#on attribute define the column through which we want to map to tables
```

```python
pd.merge(data1,data2,on='key1')  #key should be present in both the tables
```

```python
#join - it maps to tables on the basis of indexes
data1 = pd.DataFrame({'key1':[1,2,4,5,6],
                      'key2':[4,5,6,7,8],
                      'key3':[3,4,5,6,6]},
                     index = ['a','b','c','d','e']
)
data2 = pd.DataFrame({'key11':[1,2,4,5,6],
                      'key22':[4,5,6,7,8],
                      'key33':[3,4,5,6,6]
},index=['a','b','h','i','j']
)
```

```python
data1.join(data2) #by default it left outer join that is it takes all the values from the left table
```

```python
data1.join(data2,how='right')
```

```python
data1.join(data2,how='inner')
```

```python
data1.join(data2,how='outer')
```

```python
data1.join(data2,how='cross')
```
