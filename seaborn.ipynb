```python
import seaborn as sb
```

```python
sb.load_dataset('iris')
```

```python
a =sb.load_dataset('iris')
```

```python
sb.scatterplot(x =a.sepal_length,y=a.sepal_width) #we are just finding relation between columns through scatter plot
```

```python
sb.scatterplot(x =a.sepal_length,y=a.petal_length) #we are just finding relation between columns through scatter plot
```

```python
sb.scatterplot(x =a.petal_width,y=a.petal_length) #we are just finding relation between columns through scatter plot
```

```python
sb.displot(a['petal_length']) #y axis - frequency of the data, x axis - the data
```

```python
sb.load_dataset('tips')
```

```python
b =sb.load_dataset('tips')
```

```python
sb.scatterplot(x = b.total_bill,y=b.tip)
```

```python
sb.scatterplot(x = b.sex,y=b.tip)
```

```python
b['smoker'].value_counts()
```

```python
sb.relplot(x = b.total_bill,y=b.tip,data=b,hue='smoker')
```

```python
sb.relplot(x = b.total_bill,y=b.tip,data=b,hue='size') #hue uses different color shades for differentiating
```

```python
sb.relplot(x = b.total_bill,y=b.tip,data=b,style='size') #style uses different symbols  for differentiating
```

```python
sb.relplot(x = b.total_bill,y=b.tip,data=b,style='size',hue='time') #both style and hue can be used at the same time
```

```python
sb.catplot(y =b.total_bill,x= b.day) #categorical plot
```

```python
sb.jointplot(x=b.tip,y=b.total_bill) #joint plot shows the bar(frequency) as well as scatterplot of the data
```

```python
sb.pairplot(b) #pair plot takes all the numerical column and relates them with scatter plot for different columns and bar graph for same column
```

```python
sb.pairplot(a) #pair plot takes all the numerical column and relates them with scatter plot for different columns and bar graph for same column
```

```python
sb.histplot(x=b.tip,y=a.sepal_length)
```

```python
sb.barplot(x =a['petal_length'],y= a.petal_width) #y axis - frequency of the data, x axis - the data
```

```python
sb.boxplot(x=b.tip)
```
