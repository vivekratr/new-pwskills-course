```python
Q1. Explain why we have to use the Exception class while creating a Custom Exception.
Ans-> 1)When creating a custom exception, it is important to use the Exception class as the base class for your custom exception. 
2) This is because the Exception class provides a standard interface and set of behaviors that make it easier to work with and catch exceptions.
3) Since all exceptions in Python inherit from the Exception class, using it as the base class for your custom exception
allows you to catch all exceptions by using the "except Exception" statement in a try-except block.
4)This can be useful when you want to handle any type of exception in the same way.
```

```python
# Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.
# Ans->The ArithmeticError class is a built-in class in Python that is used to represent errors that occur during arithmetic operations. 
# This class is a subclass of the Exception class and is used to handle errors that are related to arithmetic operations such as
# division by zero, overflow errors, and other related issues.
# example 1: ValueError 
import logging
import math
logging.basicConfig(filename='13Feb2.log',level = logging.info)
try:
    x = -5
    y = math.sqrt(x)  
except ValueError :
    logging.error('This Value error has arrived')
    print('Value error has been handled')
except Exception as e:
    logging.error(e)
```

```python
# Example 2- ZerodivisionError
try:    
    a = 1
    b = 0
    c = a/b
except ZeroDivisionError:
    logging.error('Zero division error has been handled')
    print('Zero division error has been handled')
except Exception as e:
    logging.error(e)
```

```python
# Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.
# Ans->a) The LookupError class is used to represent errors that occur when an invalid key or index is used to access 
# a value in a sequence or mapping object.
# b) This class is used to handle errors that are related to lookup operations such as indexing, slicing, and key lookups.
try:
    e = {'a':1,'aa':2}
    x = e['b']
except KeyError:
    logging.error('Key error has been handled.')
    print('Key error has been handled.')
except Exception as e:
    print(e)
    logging.error(e)
```

```python
try:
    a = [1,2,3,4]
    b = a[6]
except IndexError:
    logging.error('Index error has been handled')
    print('Index error has been handled.')
except Exception as e:
    print(e)
    logging.error(e)
```

```python
Q6. List down some best practices for exception handling in python.
Ans->1) Always use specific errors in 'except' block which are likely to get trigger inside the try block.
2) Use multiple except block to address all the error that could arise.
3) Use Exception class at the end of all except block, to ensure that all errors will get handled.
4) Use logging instead of print in order to handle the error as it will remain even after the execution of the program.
5) Use finally block for execution important code that should be executed in any circumstances i.e closing files.
```
